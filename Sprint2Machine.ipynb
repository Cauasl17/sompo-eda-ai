```python
# Bibliotecas para manipulação e visualização de dados

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Criando dataset simulado coerente com equipamentos agrícolas

np.random.seed(42)

n = 1000  # número de registros

df = pd.DataFrame({

    # temperatura do motor (quanto maior, mais desgaste)
    "temperatura_motor": np.random.normal(85, 15, n),

    # vibração da máquina (alto indica problema mecânico)
    "vibracao": np.random.normal(50, 20, n),

    # horas de uso acumuladas
    "horas_uso": np.random.randint(100, 5000, n),

    # consumo de combustível
    "consumo_combustivel": np.random.normal(15, 5, n),

    # umidade do ambiente agrícola
    "umidade_ambiente": np.random.randint(30, 95, n),

    # idade da máquina em anos
    "idade_maquina": np.random.randint(1, 20, n),

    # manutenção atrasada (0 = não, 1 = sim)
    "manutencao_atrasada": np.random.randint(0, 2, n)
})

# Criando regra de risco (lógica realista)
df["risco_sinistro"] = (
    (df["temperatura_motor"] > 95).astype(int) +
    (df["vibracao"] > 70).astype(int) +
    (df["horas_uso"] > 3000).astype(int) +
    (df["manutencao_atrasada"] == 1).astype(int)
)

# Converter risco em 0 e 1
df["risco_sinistro"] = (df["risco_sinistro"] > 1).astype(int)

df.head()
```

```python
# CRISP-DM - Visão geral

# 1. Entendimento do negócio:
# Prever risco de falhas em equipamentos agrícolas usando IA.

# 2. Entendimento dos dados:
# Dados simulam sensores de máquinas e histórico de uso.

# 3. Preparação dos dados:
# Ajuste de valores e criação de variável alvo (risco).

# 4. Modelagem:
# Futuramente será aplicado modelo de classificação.

# 5. Avaliação:
# Identificação de variáveis mais importantes para risco.
```

```python
sns.countplot(x="risco_sinistro", data=df)
plt.title("Distribuição de Risco de Sinistro")
plt.show()
```

```python
sns.boxplot(x="risco_sinistro", y="temperatura_motor", data=df)
plt.title("Temperatura do Motor vs Risco")
plt.show()
```

```python
sns.boxplot(x="risco_sinistro", y="vibracao", data=df)
plt.title("Vibração vs Risco de Falha")
plt.show()
```

```python
sns.boxplot(x="risco_sinistro", y="horas_uso", data=df)
plt.title("Horas de Uso vs Risco")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlação entre variáveis")
plt.show()
```

```python
# Insights iniciais:

# - Máquinas com alta temperatura apresentam maior risco
# - Vibração elevada indica falhas mecânicas
# - Horas de uso impactam diretamente desgaste do equipamento
# - Manutenção atrasada é um dos principais fatores de risco
# - O modelo pode prever sinistros antes da falha acontecer
```
